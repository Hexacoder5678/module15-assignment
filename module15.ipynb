2.   Fork the repoistory :

*   click the fork button on the project's github page
*   this creates your own copy of the repository under your github account

3.   Clone the fork to your local machine : this divides your fork so you can work locally.

4.   Create a new branch : helps keep your changes orgnaized and separate from the main code

5.   Make changes :

*   edit the code, documentation,or files as needed
*   test your changes locally to make sure they work

6.  Stage ad commit changes : the commit message should clearly describe what you've done

7.  Push changes to your work : sends the changes from local->your fork in github

8. Create a pull request :

*   go to your fork on the github

*   click compare and pull request
*   provide aclear title and description for your changes


*   submit the pr to the original repository

9.  Collaborate and respond to feedback :

*   the projecy maintians may review your pr and request changes
*   make changes in the smae branch,commit and push again-the pr will update automatically.

10.  Merge and celebrate :

*   once approved,the maintainers will merge your pr into the main project
